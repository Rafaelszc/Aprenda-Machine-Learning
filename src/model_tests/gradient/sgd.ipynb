```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import pickle as pkl
from os.path import join
```

```python
DATA_PATH=join("..", "..", "..", "resources", "database", "transformed_data.pkl")

with open(DATA_PATH, "rb") as file:
    x_train, x_test, y_train, y_test= pkl.load(file)
```

```python
model = SGDRegressor(max_iter=1000, alpha=0.00001)

model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)

predict
```

```python
mse = mean_squared_error(y_test, predict)
mae = mean_absolute_error(y_test, predict)
r2 = r2_score(y_test, predict)

print(f"""
MSE: {mse:.2f}
MAE: {mae:.2f}
R2 Score: {r2:.2f}
""")
```

```python
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, predict, label="prediÃ§Ã£o")
plt.title("ComparaÃ§Ã£o dados originais")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show()
```
