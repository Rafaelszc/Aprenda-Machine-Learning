# AutoGluon

## Macros

### Imports

```python
import matplotlib.pyplot as plt
from autogluon.tabular import TabularPredictor, TabularDataset
from sklearn.model_selection import train_test_split
from os.path import join
import pandas as pd
```

### Defines

```python
DATA_PATH=join("..", "..", "..", "resources", "database", "Vegetable_market.csv")

data = pd.read_csv(DATA_PATH)

train, test = train_test_split(data, test_size=0.25, random_state=0)
```

```python
label = "Price per kg"
```

## Data Manipulation

### Model Fit

```python
predictor = TabularPredictor(
            label=label, problem_type="regression", 
            eval_metric="root_mean_squared_error")
```

```python
predictor.fit(
            train, presets="best_quality",
            # hyperparameters={}
            )
```

```python
predictor.leaderboard(test, silent=True)
```

### Predict

```python
predict = predictor.predict(test)
```

```python
outcome = pd.DataFrame(data={
    "target": test["Price per kg"],
    "predict": predict
    }
)

outcome
```

```python
x_ax = range(len(test[label]))
plt.plot(x_ax, test[label], label="original")
plt.plot(x_ax, predict, label="predição")
plt.title("Comparação dados originais")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show()
```
