```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import pickle as pkl
from os import path
```

```python
DATA_PATH = path.join("..", "..", "resources", "database", "Vegetable_market.csv")

data = pd.read_csv(DATA_PATH)
data
```

```python
data_x = data.iloc[:, :-1]

data_x
```

```python
data_y = data.iloc[:, -1]

data_y = data_y.values.reshape(-1, 1)

data_y
```

```python
data_x.columns
```

```python
x_categorical = ['Vegetable', 'Season',
                'Month', 'Deasaster Happen in last 3month',
                'Vegetable condition']

x_numerical = ["Temp"]
```

```python
def encode_data(data: pd.Series) -> pd.DataFrame:
    encoder = LabelEncoder()
    data_encoded = encoder.fit_transform(data)

    return data_encoded
```

```python
def scale_data(data: pd.Series) -> pd.DataFrame:
    scaler = StandardScaler()

    data_scaled = scaler.fit_transform(data)

    return data_scaled
```

```python
def one_hot_encoding(data: pd.DataFrame) -> pd.DataFrame:
    encoder = OneHotEncoder(sparse_output=False)

    data_encoded = encoder.fit_transform(data)

    df = pd.DataFrame(data=data_encoded, columns=encoder.get_feature_names_out())

    return df
```

```python
x_encoded = data_x[x_categorical].apply(lambda x: encode_data(x))

x_encoded
```

```python
x_encoded = one_hot_encoding(x_encoded)

x_encoded
```

```python
temp = data_x[x_numerical].values.reshape(-1, 1)

temp_scaled = scale_data(temp)

temp_scaled
```

```python
data_x_transformed = x_encoded

data_x_transformed.insert(3, "Temp", temp_scaled)

data_x_transformed
```

```python
data_y_scaled = scale_data(data_y)

data_y_scaled
```

```python
trans_x_train, trans_x_test, trans_y_train, trans_y_test = train_test_split(
                                                        data_x_transformed, data_y_scaled, 
                                                        test_size=0.25, random_state=0)
```

```python
TRANSFORMED_DB_PATH = path.join("..", "..", "resources", "database", "transformed_data.pkl")

with open(TRANSFORMED_DB_PATH, "wb") as file:
    pkl.dump([trans_x_train, trans_x_test, trans_y_train, trans_y_test], file)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
                                    data_x, data_y,
                                    test_size=0.25, random_state=0)
```

```python
DEFAULT_DB_PATH = path.join("..", "..", "resources", "database", "default_data.pkl")

with open(DEFAULT_DB_PATH, "wb") as file:
    pkl.dump([x_train, x_test, y_train, y_test], file)
```
